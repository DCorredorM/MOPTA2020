```python
from scipy.stats import chisquare
print(chisquare(f_obs=p_obs,f_exp=p_dist))
print('No se puede aceptar la hipÃ³tesis nula de que la distribuciÃ³n empÃ­rica')
print('se ajusta a una distribuciÃ³n.')
dif = sum((np.array(p_obs) - np.array(p_dist))**2/(np.array(p_dist)))
plt.bar(range(0,len(p_obs)),p_dist)
plt.bar(range(0,len(p_obs)),p_obs,color='r')
plt.bar(range(0,len(p_obs)),np.array(p_obs) - np.array(p_dist),color='g')
plt.legend(['Observed','Expected','Diference'])
plt.xlabel('Number of units in an order')
plt.ylabel('Relative frecuency')
plt.show()
```

```python
from scipy.stats import poisson
obs = np.array(numbers)/len(sample)
dif = {}
mu_sample = sample.mean()

for mu in np.arange(mu_sample-2,12,.1):
    rv = np.array([poisson.pmf(x, mu) for x in range(1,max(sample)+1)])
    dif[mu] = np.sum((rv - obs)**2/rv)
    
best = min(dif.keys(), key=(lambda k: dif[k]))
print('El ajuste se realiza con nÃºmero de Ã©xitos por unidad de',best)
p_dist = np.array([poisson.pmf(x, best) for x in range(1,27+1)])
    
plt.plot(p_dist,color='r')
plt.bar(range(1,max(sample)+1),np.array(numbers)/len(sample))
plt.xlabel('Number of units per order')
plt.ylabel('Relative frecuency')
plt.legend(['Poisson','Observed'])
plt.show()
```

# Cluster by distance

```python
#Import packages
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.cluster import Birch
```

```python
#1. Import daily demmand
data_path = os.chdir("../Data") #Data path
Demands2018=import_data('mopta2020_q2018.csv',names=['id']+list(range(1,366))).set_index('id')
Demands2019=import_data('mopta2020_q2019.csv',names=['id']+list(range(1,366))).set_index('id')
Demands2018.head()
```

```python
#2. Plot latitudes and longitudes
class data_reader():
	"""docstring for data_reader"""
	def __init__(self):
		super(data_reader, self).__init__()
		self.path = os.chdir("../Data")
	def import_data(self,file):
		return pd.read_csv(file,header=None)


df=data_reader().import_data('mopta2020_vertices.csv')
df.rename(columns={0: "id", 1: "loc_name",2:'long',3:'lat'},inplace=True)
print(df.head())

df.plot.scatter(x='long',y='lat')
plt.show()
```

```python
#3. Clustering by Kmeans and Birch
long = np.array(df['long'])
lat = np.array(df['lat'])
coord = np.column_stack((long,lat))
# Create a KMeans instance with 8 clusters: model
model = KMeans(n_clusters=8)
# Fit model to points
model.fit(coord)
labels = model.predict(coord)
plt.scatter(coord[:,0],coord[:,1],c=labels, alpha=0.5, cmap='rainbow')
plt.show()
# Create a Birch instance with 8 clusters: model
model = Birch(threshold=0.01, n_clusters=8)
# Fit model to points
model.fit(coord)
labels = model.predict(coord)
plt.scatter(coord[:,0],coord[:,1],c=labels, alpha=0.5, cmap='rainbow')
plt.show()
```

```python
ks = range(1, 20)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters = k)
    
    # Fit model to samples
    model.fit(coord)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
