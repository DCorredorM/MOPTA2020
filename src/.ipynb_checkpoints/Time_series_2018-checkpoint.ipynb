# Analysis of demand throughout time (2018)

We are going to do a first analysis as follows: 

1. Import the data of the daily demand of all the clients 
2. Agregate demand of all the clients for any day
3. Plot the demand

```python
#Import packages
import os
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf #For the time-series ACF
from statsmodels.tsa.stattools import adfuller,kpss    #For stationarity assessment
from statsmodels.tsa.arima_process import ArmaProcess #To simulate ARMA models
from statsmodels.tsa.arima_model import ARMA,ARIMA #To fit ARMA models
from statsmodels.tsa.statespace.sarimax import SARIMAX
import statsmodels.api as sm #Para regresiones lineales 
```

```python
#Define functions
def import_data(file,names,h=None):
    '''
    Imports data as a DataFrame receiving the file to import in csv format
    and the name that the columns will have in the DataFrame.
    '''
    return pd.read_csv(file,header=h,names=names)
def plot_dem_timeseries(Demands, client):
    '''
    Plots the demand of one client as a time series.
    '''
    plt.plot(Demands.columns, Demands.loc[client,:])
def line_format(label):
    """
    Convert time label to the format of pandas line plot
    """
    month = label.month_name()[:3]
    if month == 'Jan':
        month += f'\n{label.year}'
    return month
```

```python
#1. Import daily demmand
data_path = os.chdir("../Data") #Data path
Demands2018=import_data('mopta2020_q2018.csv',names=['id']+list(range(1,366))).set_index('id')
Demands2019=import_data('mopta2020_q2019.csv',names=['id']+list(range(1,366))).set_index('id')
Demands2018.head()
```

```python
#2. Aggregate demand by day 
day_demands2018 = Demands2018.sum() #This is the series
day_demands2019 = Demands2019.sum()
day_demands2019.index = range(366,730+1)
day_demands = day_demands2018.append(day_demands2019)
```

```python
#3. Plot the time series
fig, ax = plt.subplots()
ax.plot(day_demands)
ax.grid(True)
ax.set_xlabel('Day')
ax.set_ylabel('Units')
ax.set_title('Aggregated daily demand')
plt.show()
```

# Look at the time series

Now we are going to create a time series object: 

1. Put dates on the time series
 1. Aggregate by week
 2. Aggregate by month 

```python
#1. Put dates on the time series
idx = pd.date_range(start='1/1/2018', end='12/31/2018', freq='D')
day_demands2018_dates = pd.DataFrame(day_demands2018, columns=['Demand']).set_index(idx)

idx = pd.date_range(start='1/1/2019', end='12/31/2019', freq='D')
day_demands2019_dates = pd.DataFrame(day_demands2019, columns=['Demand']).set_index(idx)

day_demands_dates = day_demands2018.append(day_demands2019) #This is the series of the two years
idx = pd.date_range(start='1/1/2018', end='12/31/2019', freq='D')
day_demands_dates = pd.DataFrame(day_demands, columns=['Demand']).set_index(idx)
day_demands_dates.head()
```

```python
#1.Put dates on the time series
month_demands_dates = day_demands_dates.resample(rule='M').sum()
month_demands2018_dates = day_demands2018_dates.resample(rule='M').sum()

plt.clf()
fig, ax = plt.subplots()
ax.plot(day_demands_dates)
ax.grid(True)
ax.set_xlabel('Day')
ax.set_ylabel('Units')
ax.set_title('Aggregated daily demand')
ax.set_xticks(month_demands_dates.index)
ax.set_xticklabels(map(lambda x: line_format(x), month_demands_dates.index))
plt.gcf().set_size_inches(15, 5)
plt.show()
```

```python
#A. Plot the weekly time series
week_demands_dates = day_demands_dates.resample(rule='W').sum()[:-1]
plt.clf()
fig, ax = plt.subplots()
ax.plot(week_demands_dates)
ax.grid(True)
ax.set_xlabel('Week')
ax.set_ylabel('Units')
ax.set_title('Aggregated weekly demand')
ax.set_xticks(month_demands_dates.index)
ax.set_xticklabels(map(lambda x: line_format(x), month_demands_dates.index))
plt.gcf().set_size_inches(15, 5)
plt.show()
```

```python
#B. Plot the monthly time series
plt.clf()
month_demands_dates = day_demands_dates.resample(rule='M').sum()
fig, ax = plt.subplots()
ax.plot(month_demands_dates)
ax.grid(True)
ax.set_xlabel('Month')
ax.set_ylabel('Units')
ax.set_title('Aggregated monthly demand (2018)')
ax.set_xticks(month_demands_dates.index)
ax.set_xticklabels(map(lambda x: line_format(x), month_demands_dates.index))
plt.gcf().set_size_inches(15, 5)
plt.show()
```

## Check the autocorrelation of the daily series

1. Check the autocorrelation of the daily series
2. Check if the data follows a random walk movement

```python
#1. Check the analytic autocorrelation 
day_demands_dates_diff = day_demands_dates.diff()
day_demands_autocorr = day_demands_dates_diff['Demand'].autocorr()
print('The data has a autocorrelation coefficient of', day_demands_autocorr)
print('There is a negative autocorrelation in the data with the first lag (lag 1)')
```

```python
#1. Check the ACF and the PACF graph
fig, axes = plt.subplots(2,1)
plot_acf(day_demands_dates,lags=25,alpha=0.05,title='Autocorrelation (ACF) daily demand (2018)', ax = axes[0])
plot_pacf(day_demands_dates,lags=25,alpha=0.05,title='Partial autocorrelation (PACF) daily demand (2018)', ax = axes[1])
plt.tight_layout()
plt.show()
```

```python
#2. Check whether the data follows a random walk 
#Of the data itself 
day_demands_adfuller = adfuller(day_demands_dates)
print('The p-value of the test on demands with adfuller is:',day_demands_adfuller[1])
print('Since the p-value is lower than 0.05 we cannot reject the alternative hipothesis that the data is stationary\n')

day_demands2018_adfuller = adfuller(day_demands2018_dates)
print('The p-value of the test on demands in 2018 with adfuller is:',day_demands2018_adfuller[1])
print('Since the p-value is greater than 0.05 we cannot reject the null hipothesis that the data is *not* stationary\n')

day_demands2019_adfuller = adfuller(day_demands2019_dates)
print('The p-value of the test on demands in 2019 with adfuller is:',day_demands2019_adfuller[1])
print('Since the p-value is greater than 0.05 we cannot reject the null hipothesis that the data is *not* stationary\n')
```

```python
#Correct non-stationary data 
#1. check trend with a linear regression 
time = pd.Series(day_demands2018.index.astype(float))
day_demands2018_y = pd.DataFrame(day_demands2018)
day_demands2018_y.index = time.index
time2018_x = sm.add_constant(time)
result = sm.OLS(day_demands2018_y, time2018_x).fit()
print(result.summary())
print('The p-value of the F-statistic of the fully specified model is', result.f_pvalue,'. Since it is above 0.05 we can say the data has no trend.')
```

```python
#Correct non-stationary data 
#2. Check the seasonality
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(day_demands2018_dates, model='additive')
#plt.figure(figsize=(10,10)) 
result.plot()
plt.gcf().set_size_inches(15, 5)
plt.show()
print('We can see in the seasonal descomposition that we have seasonal data.\n The resid tell us that the data seems to have constant standard deviation ')
```

```python
#3. Correct seasonality
lag1 = day_demands2018_dates.Demand - day_demands2018_dates.Demand.rolling(7).mean() 
lag1 = lag1.dropna()
fig, axes = plt.subplots(2,1)
plot_acf(lag1,lags=25,alpha=0.05,title='Autocorrelation (ACF) daily demand diff 7 (2018)', ax = axes[0])
plot_pacf(lag1,lags=25,alpha=0.05,title='Partial autocorrelation (PACF) daily demand diff 7 (2018)', ax = axes[1])
plt.tight_layout()
plt.show()

plt.clf()
lag1 = lag1 - lag1.shift(1)
lag1 = lag1.dropna()
fig, axes = plt.subplots(2,1)
plot_acf(lag1,lags=25,alpha=0.05,title='Autocorrelation (ACF) daily demand diff 7 + diff 1 (2018)', ax = axes[0])
plot_pacf(lag1,lags=25,alpha=0.05,title='Partial autocorrelation (PACF) daily demand diff 7 + diff 1 (2018)', ax = axes[1])
plt.tight_layout()
plt.show()

result = seasonal_decompose(lag1, model='additive')
#plt.figure(figsize=(10,10)) 
result.plot()
plt.gcf().set_size_inches(15, 5)
plt.show()

plt.clf()
fig, ax = plt.subplots()
ax.plot(lag1)
ax.grid(True)
ax.set_xlabel('Day')
ax.set_ylabel('Units')
ax.set_title('Aggregated daily demand')
ax.set_xticks(month_demands2018_dates.index)
ax.set_xticklabels(map(lambda x: line_format(x), month_demands2018_dates.index))
plt.gcf().set_size_inches(15, 5)
plt.show()
```

```python
lag1_adfuller = adfuller(lag1)
print('The p-value of the test on demands is:',lag1_adfuller[1])
print('Since the p-value is lower than 0.05 we cannot reject the alternative hipothesis that the data is stationary\n')
```

```python
#idx = pd.date_range(start='1/1/2018', end='12/31/2018', freq='D')
#day_demands2018_dates = pd.DataFrame(day_demands2018, columns=['Demand']).set_index(idx)
AIC = {}
for p in range(3):
    for q in range(3):
        try: 
            sarima_model= SARIMAX(lag1, order=(p, 1, q), seasonal_order=(p, 1, q, 7), enforce_invertibility=False, enforce_stationarity=False)
            res = sarima_model.fit()
            AIC['AR:'+str(p)+',MA:'+str(q)] = res.aic
        except: 
            print('No lo logrÃ©')
            pass
        
# Plot the AIC as a function of p
plt.bar(list(AIC.keys()), list(AIC.values()))
plt.xticks(list(AIC.keys()))
plt.xlabel('Order of ARIMA Model')
plt.ylabel('Akaike Information Criterion')
plt.gcf().set_size_inches(15, 5)
plt.show()
print(AIC)
```

```python
train_st = day_demands2018_dates
test_st = day_demands2019_dates[0:10]
sarima_model= SARIMAX(lag1, order=(0, 1, 1), seasonal_order=(1, 1, 1, 7), enforce_invertibility=False, enforce_stationarity=False)
sarima_fit = sarima_model.fit()
sarima_pred = sarima_fit.forecast(10)
sarima_pred = pd.Series(sarima_pred)
predicted_intervals = sarima_pred.conf_int(alpha=0.05)
lower_bounds = predicted_intervals['lower station'] + day_demands2018_dates.Demand.iloc[-16:].rolling(7).mean().dropna().values
upper_bounds = predicted_intervals['upper station'] + day_demands2018_dates.Demand.iloc[-16:].rolling(7).mean().dropna().values
sarima_pred.index = test_st.index

df = pd.concat([test_st,sarima_pred], axis=1)
df.columns = ['Read','Forecast']
df_sum = df['Forecast'] + day_demands2018_dates.Demand.iloc[-16:].rolling(7).mean().dropna().values
df = pd.concat([df,df_sum],axis=1)
df.columns = ['Real','Forecast','Adj forecast']
print(df)

sarima_rmse = np.sqrt(np.mean(np.square(test_st.values - df['Adj forecast'].values)))

plt.clf()
fig, ax = plt.subplots(figsize=(15, 4))
ax.plot(day_demands2018_dates.index, day_demands2018_dates.values)
ax.plot(test_st.index, test_st.values, label='truth', color = 'green')
ax.plot(test_st.index, df['Adj forecast'].values, color='#ff7823', linestyle='--', label="RMSE={:0.2f}".format(sarima_rmse))
ax.legend()
ax.set_title("SARIMA")
```
